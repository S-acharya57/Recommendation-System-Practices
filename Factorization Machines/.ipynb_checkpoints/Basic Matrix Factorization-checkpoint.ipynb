# Matrix Factorization

- sparse user-item matrix factorized into approx product of 2 low-rank matrices
    - two matrices contain user and item embeddings

```python
import pandas as pd
import numpy as np 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt 
```

```python
user_ratings = pd.read_csv('user_movie_matrix.csv')
user_ratings = user_ratings.fillna(0)
user_ratings.head(2)
```



## Assumptions of MF:
    - each user can be described by 'd' features
    - movie by 'd' set of analogous features 

$$ \huge L = \sum (r_ui - x_uy^T_i)^2 + \lambda(\sum|x_u||^2 + \sum|y_i||^2 )  $$

## Some functions for matrix factorization process









```python
class MF:
    def __init__(self, iters, num_factors, reg):
        """
        reg: Regularization Term 
        """
        self.iters = iters
        self.num_factors = num_factors 
        self.reg = reg

    def fit(self, train, test):
        self.num_users, self.num_movies = train.shape
        self.user_factors = np.random.random((self.num_users, self.num_factors))
        self.movie_factors = np.random.random((self.num_movies, self.num_factors))

        self.train_errors = []
        self.test_errors = []

        for i in range(self.iters):
            self.user_factors = self._als_step(train, self.user_factors, self.movie_factors)
            self.movie_factors = self._als_step(train.T, self.movie_factors, self.user_factors)

            predictions = self.predict()
            # print(f'Predictions shape{predictions.shape}\nTrain shape:{train.shape}')
            train_error = self.mse_score(train, predictions)
            test_error = self.mse_score(test, predictions)

            self.train_errors.append(train_error)
            self.test_errors.append(test_error)

        return self 

        
    # alternating least squares
    # fixed vec and solve_vec fluctuates between U and I matrix
    # this is loss function derivative to get each step solution
    def _als_step(self, ratings, solve_vec, fixed_vec):
        """
        Fixed vec is user or movie matrix while updating the other one
        """
        # print(f'Ratings shape:{ratings.shape}\nSolve_vec:{solve_vec.shape}\n, Fixed vec:{fixed_vec.shape}')
        A = fixed_vec.T.dot(fixed_vec) + np.eye(self.num_factors) * self.reg
        b = ratings.dot(fixed_vec)
        A_inv = np.linalg.inv(A)
        solve_vec = b.dot(A_inv)
        return solve_vec

    def predict(self):
        pred = self.user_factors.dot(self.movie_factors.T)
        return pred 

    @staticmethod
    def mse_score(y_true, y_pred):
        "ignoring 0 terms for the sake of actual errors computed"
        mask = np.nonzero(y_true)
        mse = mean_squared_error(y_true[mask], y_pred[mask])
        return mse 
```

```python
# converting to np array
user_ratings_arr = user_ratings.to_numpy()
user_ratings_arr, user_ratings_arr.shape, user_ratings.shape
```

```python
als = MF(iters=50, num_factors = 50, reg=0.001)
als 
```

## 3. Creating train-test for ratings 


```python
def train_test(data_arr):
    test = np.zeros(data_arr.shape)
    train = data_arr.copy() 
    
    for user in range(data_arr.shape[0]):
        #flatnonzero returns indices that are non-zero in the flattened version of matrix
        test_index = np.random.choice(np.flatnonzero(data_arr[user]), size=10, replace=False)

        # keeping the test_index in data array as 0
        # so that it can be solved for test case scenario
        train[user, test_index] = 0.0
        test[user, test_index] = data_arr[user, test_index] 

    assert np.all(train*test ==0)
    return train, test
```

```python
train, test = train_test(user_ratings_arr)
train.shape, test.shape
```

```python
als.fit(train, test)
als 
```

```python
def loss_curve(model):
    "Loss curves for train and test"
    linewidth = 3
    plt.plot(model.test_errors, label = 'Test', linewidth = linewidth)
    plt.plot(model.train_errors, label = 'Train', linewidth = linewidth)
    plt.xlabel('iterations')
    plt.ylabel('MSE')
    plt.legend(loc = 'best')
```

```python
# with use of all rows in user-movie matrix in ratings_small.csv 
loss_curve(als)
```
